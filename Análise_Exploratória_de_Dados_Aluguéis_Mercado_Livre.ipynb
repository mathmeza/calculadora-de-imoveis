```python
# fazer histograma de IQR do preco
# pegar outliers
# fazer histograma com base logaritima 
```

```python
df_filtrado.describe().T
```

```python
# Calculo IQR da área
q1_area = df['area'].quantile(0.25)
q3_area = df['area'].quantile(0.75)
IQR_area = q3_area - q1_area

# Calculo IQR do preço
q1_preco = df['preco'].quantile(0.25)
q3_preco = df['preco'].quantile(0.75)
IQR_preco = q3_preco - q1_preco

# Calculo IQR de quartos
q1_quartos = df['quartos'].quantile(0.25)
q3_quartos = df['quartos'].quantile(0.75)
IQR_quartos = q3_quartos - q1_quartos
```

```python
# Podemos utilizar a fórmula do IQR para pegar os Outliers (Conforme visto nos boxplots)
lower_outlier_quartos = max((q1_quartos-1.5*IQR_quartos),0)
upper_outlier_quartos = (q3_quartos+1.5*IQR_quartos)

lower_outlier_area = max((q1_area-1.5*IQR_area),0)
upper_outlier_area = (q3_area+1.5*IQR_area)

lower_outlier_preco = max((q1_preco-1.5*IQR_preco),0)
upper_outlier_preco = (q3_preco+1.5*IQR_preco)


print(f'Outlier lower da variável quartos: {lower_outlier_quartos} ')
print(f'Outlier upper da variável quartos: {upper_outlier_quartos} ')
print('---'*30)
print(f'Outlier lower da variável area: {lower_outlier_area}')
print(f'Outlier upper da variável area: {upper_outlier_area}')
print('---'*30)
print(f'Outlier lower da variável preco: {lower_outlier_preco}')
print(f'Outlier upper da variável preco: {upper_outlier_preco}')
```
